#  Задание 1
1. Скачать данные

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('kc_house_data.csv')
df
```

2. Изучите стоимость недвижимости

```python
plt.figure(figsize=(12, 6))
plt.hist(df['price'], bins=32, color='lightgrey', edgecolor='black')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.grid()
plt.show()
```

3. Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(12, 6))
plt.hist(df['sqft_living'], bins=20, color='lightgrey', edgecolor='black')
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество')
plt.grid()
plt.show()
```

4. Изучите распределение года постройки

```python
plt.figure(figsize=(12, 6))
plt.hist(df['yr_built'], bins=50, color='lightgrey', edgecolor = 'black')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
plt.grid()
plt.show()
```

Рассматривая данные графики можно сделать следующие выводы:
* Гистограмма распределения стоимости недвижимости имеет ярко выраженный пик в области 300000 - 500000, это может горворить о востребованности недвижимости в этом ценовом диапазоне.


* Форма гистограммы распределения квадратуры жилой площади дает представление о средней жилой площади около 2000.


* По гистограмме распределения года постройки можно проследить за количеством построек в разные годы и выявить периоды строительной активности в начале 1950-х, конце 70-х и, особенно, в 2010.

# Задание 2
1. Изучите распределение домов от наличия вида
на набережную

```python
plt.figure(figsize=(6, 6))
plt.pie(df['waterfront'].value_counts(), labels=['Без вида', 'С видом на набережную'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgrey'])
plt.title('Распределение домов в зависимости от вида на набережную')
plt.show()
```

2. Изучите распределение этажей домов

```python
plt.figure(figsize=(12, 6))
df['floors'].value_counts().sort_index().plot(kind='bar', color='grey')
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.xticks(rotation=0)
plt.show()
```

3. Изучите распределение состояния домов

```python
plt.figure(figsize=(12, 6))
df['condition'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.xticks(rotation=0)
plt.show()
```

Визуализация распределения домов в зависимости от вида на набережную позволяет сделать следующие выводы. Большинство домов не имеют вид на набережную, в то время как дома с видом на набережную представляют собой совсем небольшую долю.

Анализ распределения этажей домов дает нам следующую картину. Больше всего домов с 1 этажом, затем идут дома с 2 этажами, 1,5 этажами, 3 этажами и 2,5 этажами. Эти данные могут быть полезны при планировании строительных проектов.

Изучив распределение состояния домов видно, что большинство домов имеют состояние на уровне 3. Эта информация может быть важной при принятии решений о покупке или продаже недвижимости, так как состояние дома может влиять на его стоимость.

# Задание 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
plt.figure(figsize=(12, 6))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5)
plt.title('Влияние площади жилой площади на стоимость')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость')
plt.show()
```

На графике рассеяния видно, что с увеличением жилой площади стоимость имеет тенденцию увеличиваться. Это говорит о том, что площадь жилья оказывает влияние на цену недвижимости.

```python
plt.figure(figsize=(6, 6))
sns.heatmap(df[['sqft_living', 'bedrooms', 'grade', 'floors', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

На тепловой карте корреляций видно, что есть некоторые взаимосвязи между различными характеристиками.

Оценка и жилая площадь: присутствует сильная положительная корреляция, это означает, что дома с бОльшей площадью оцениваются выше.

Оценка и стоимость: наблюдается положительная корреляция, что подтверждает предположение, что более высокая оценка соответствует более высокой стоимости недвижимости.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data = df, x = 'yr_built', y = 'price')
plt.title('Влияние года постройки на стоимость')
plt.xlabel('Год')
plt.ylabel('Стоимость')
plt.show();
```

График показывает, как стоимость недвижимости распределена в зависимости от года постройки. Из графика видно, что:

Существует разнообразие цен для домов разного возраста. Некоторые старые дома имеют высокую стоимость, возможно, из-за их исторической ценности или редких характеристик.

Дома, построенные в более поздние годы, имеют более высокую стоимость в среднем. Это может быть связано с современными стандартами строительства, улучшенными удобствами или модными архитектурными решениями.

Несмотря на общий тренд, видны выбросы, где дома старых построек имеют высокую стоимость, а современные дома могут иметь низкую. Возможно, это связано с уникальными характеристиками или редкостью таких домов.

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='long', y='lat', hue='grade', palette='viridis', data=df)
plt.title('Распределение оценок в зависимости от координат')
plt.xlabel('Долгота')
plt.ylabel('Широта')
plt.show()
```

График распределения оценки домов в зависимости от координат (широты и долготы) предоставляет следующие наблюдения:

На графике видны области с различными оценками недвижимости. Некоторые районы могут быть характеризованы высокими оценками, что может свидетельствовать о высоком качестве строений или благоприятной инфраструктуре.

Некоторые места имеют выбросы в оценках, что может свидетельствовать о некоторых уникальных особенностях или премиум-характеристиках в этих районах.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='floors', y='view', data=df)
plt.title('Зависимость оценки вида от этажности дома')
plt.xlabel('Этажность дома')
plt.ylabel('Оценка вида')
plt.show()
```

Видно, что более высокие этажи обычно имеют более высокие оценки вида. Это может быть связано с тем, что дома на более высоких этажах обеспечивают лучший обзор и вид на окружающую местность.

В некоторых случаях, даже на более низких этажах, может встречаться высокая оценка вида, что может быть связано с особенностями расположения дома или окружающей местности.
